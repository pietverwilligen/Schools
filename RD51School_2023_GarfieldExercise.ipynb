# Ready to start using Garfield!

```python
# Set up the gas medium.
gas = ROOT.Garfield.MediumMagboltz()
gas.SetTemperature(293.15)
gas.SetPressure(740.)
gas.SetComposition("ar", 80., "co2", 20.)
gas.EnableDrift()
# Load the electron-molecule cross-sections.
gas.Initialise(True)
```

```python
gas.PlotElectronCrossSections()
ROOT.gROOT.GetListOfCanvases().FindObject('cCs_0').Draw()
ROOT.gROOT.GetListOfCanvases().FindObject('cCs_1').Draw()
```

```python
# Set the Penning transfer probability.
rPenning = 0.51
gas.EnablePenningTransfer(rPenning, 0., "ar")
```

```python
# Load the field map.
fm = ROOT.Garfield.ComponentAnsys123()
fm.Initialise("GEM-FEM-FieldMaps/ELIST.lis", "GEM-FEM-FieldMaps/NLIST.lis", "GEM-FEM-FieldMaps/MPLIST.lis", "GEM-FEM-FieldMaps/PRNSOL.lis", "mm")
fm.EnableMirrorPeriodicityX()
fm.EnableMirrorPeriodicityY()
fm.SetGas(gas)
```

```python
# Dimensions of the GEM [cm].
pitch = 0.014
```

```python
fieldView = ROOT.Garfield.ViewField()
cF = ROOT.TCanvas('cF', '', 600, 600)
fieldView.SetCanvas(cF)
fieldView.SetComponent(fm)
# Set the viewing plane (xz plane).
fieldView.SetPlaneXZ()
# Set the plot limits in the current viewing plane.
fieldView.SetArea(-0.5 * pitch, -0.02, 0.5 * pitch, 0.02)
fieldView.SetVoltageRange(-160., 160.)
```

```python
cF.SetLeftMargin(0.16)
fieldView.Plot("v", "cont1z")
cF.Update()
cF.Draw()
```

```python
sensor = ROOT.Garfield.Sensor()
sensor.AddComponent(fm)
sensor.SetArea(-5 * pitch, -5 * pitch, -0.01, 5 * pitch,  5 * pitch, 0.025)
```

```python
aval = ROOT.Garfield.AvalancheMicroscopic()
aval.SetSensor(sensor)

drift = ROOT.Garfield.AvalancheMC()
drift.SetSensor(sensor)
drift.SetDistanceSteps(2.e-4)
```

```python
driftView = ROOT.Garfield.ViewDrift()
aval.EnablePlotting(driftView, 1)
drift.EnablePlotting(driftView)
```

```python
x0 = -0.5 * pitch + ROOT.Garfield.RndmUniform() * pitch
y0 = -0.5 * pitch + ROOT.Garfield.RndmUniform() * pitch
z0 = 0.02
t0 = 0.
e0 = 0.1
aval.AvalancheElectron(x0, y0, z0, t0, e0, 0., 0., 0.)
```

```python
for electron in aval.GetElectrons():
  p0 = electron.path[0]
  drift.DriftIon(p0.x, p0.y, p0.z, p0.t)
```

```python
cD = ROOT.TCanvas('cD', '', 600, 600)
meshView = ROOT.Garfield.ViewFEMesh()
meshView.SetArea(-2 * pitch, -0.02, 2 * pitch, 0.02)
meshView.SetCanvas(cD)
meshView.SetComponent(fm)
# x-z projection.
meshView.SetPlane(0, -1, 0, 0, 0, 0)
meshView.SetFillMesh(True)
#  Set the color of the kapton.
meshView.SetColor(2, ROOT.kYellow + 3)
meshView.EnableAxes()
meshView.SetViewDrift(driftView)
```

```python
meshView.Plot()
cD.Draw()
```
