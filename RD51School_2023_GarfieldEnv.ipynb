# Install Garfield

```python
!git clone https://gitlab.cern.ch/garfield/garfieldpp.git
!apt-get install libgsl-dev
```

```python
!mkdir garfield
!mv garfieldpp garfield
!mkdir garfield/build
!export ROOTSYS="/content/root/install"
!export PATH="$ROOTSYS/bin:$PATH"
!export LD_LIBRARY_PATH="$ROOTSYS/lib:$LD_LIBRARY_PATH"
!cd garfield/build && cmake -DWITH_EXAMPLES=OFF -DPYTHON_EXECUTABLE=/usr/bin/python3 -DPYTHON_LIBRARY=/usr/lib/x86_64-linux-gnu/libpython3.9m.so -DPYTHON_INCLUDE_DIR=/usr/include/python3.9 -DCMAKE_INSTALL_PREFIX=`pwd`/../install ../garfieldpp
!cd garfield/build && make -j 2
!cd garfield/build && make install
```

```python
!ls /content/garfield/install/lib
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the environment variables.
import os
os.environ["GARFIELD_HOME"] = "/content/garfield/garfieldpp"
os.environ["ROOTSYS"] = "/content/root/"
# os.environ["PATH"] += ":/content/elmer/install/bin:/content/gmsh/bin"

# Load the ROOT and Garfield libraries.
import sys
sys.path.append("/content/root/install/lib")

import ctypes
ctypes.cdll.LoadLibrary('/content/root/lib/libCore.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libThread.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libImt.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libRIO.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libNet.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libTree.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libMathCore.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libMatrix.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libHist.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libGeom.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libXMLIO.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libGdml.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libGraf.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libGpad.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libGraf3d.so')
ctypes.cdll.LoadLibrary('/content/root/lib/libcppyy_backend3_10.so')


import ROOT
ROOT.gSystem.Load("/content/garfield/install/lib/libmagboltz.so")
ROOT.gSystem.Load("/content/garfield/install/lib/libdegrade.so")
ROOT.gSystem.Load("/content/garfield/install/lib/libGarfield.so")
%jsroot off
```

**Now the proof that GARFIELD is working:**

```python
# Set up the gas medium.
gas = ROOT.Garfield.MediumMagboltz()
gas.SetTemperature(293.15)
gas.SetPressure(740.)
gas.SetComposition("ar", 90., "co2", 10.)
gas.EnableDrift()
```
